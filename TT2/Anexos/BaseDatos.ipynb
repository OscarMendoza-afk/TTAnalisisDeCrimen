```python
import pandas as pd
import pymysql
from pathlib import Path
```

```python
#Creacion de conexion con base de datos
connection = pymysql.connect(host='localhost',
                             user='DataCriming',
                             password='DB1202',
                             db='Crimenes', 
                             port = 3306)

#Creacion de cursor para base de datos
cursor = connection.cursor()
```

```python
df = pd.read_csv('UpdateCrimenesCleanV.csv', low_memory=False)
```
